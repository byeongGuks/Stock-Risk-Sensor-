```python
from pykrx import stock
import pandas as pd
```

### 기관별 시장 투자 현황 정보 크롤링

```python
kospi_trade_volumn = stock.get_market_trading_value_by_date("20170101", "20220930", "KOSPI", etf=True, etn=True, elw=False, detail=True)
kospi_trade_volumn_elw = stock.get_market_trading_value_by_date("20170101", "20220930", "KOSPI", etf=True, etn=True, elw=True, detail=True)

kosdaq_trade_volumn = stock.get_market_trading_value_by_date("20170101", "20220930", "KOSDAQ", etf=True, etn=True, elw=False, detail=True)
kosdaq_trade_volumn_elw = stock.get_market_trading_value_by_date("20170101", "20220930", "KOSDAQ", etf=True, etn=True, elw=True, detail=True)

print(kospi_trade_volumn.head())
```

```python
kospi_trade_volumn_elw = (kospi_trade_volumn_elw - kospi_trade_volumn)
kosdaq_trade_volumn_elw = (kosdaq_trade_volumn_elw - kosdaq_trade_volumn)
(kospi_trade_volumn_elw).head()
```

```python
kospi_trade_volumn.to_csv('./data/kospi_trade_volumn.csv', encoding = 'euc-kr')
kospi_trade_volumn_elw.to_csv('./data/kospi_trade_volumn_elw.csv', encoding = 'euc-kr')
kosdaq_trade_volumn.to_csv('./data/kosdaq_trade_volumn.csv', encoding = 'euc-kr')
kosdaq_trade_volumn_elw.to_csv('./data/kosdaq_trade_volumn_elw.csv', encoding = 'euc-kr')
```

### 종목별 투자 현황 크롤링

```python
kospi_stock_list = pd.read_csv('./data/kospi.csv')
kosdaq_stock_list = pd.read_csv('./data/kosdaq.csv')
kospi_stock_list.head()
```

```python
kospi_stock_trade_volume = pd.DataFrame()
i = 0
error_code_list = []
for code in kospi_stock_list['code'] :
    try : 
        df = stock.get_market_trading_volume_by_date("20170101", "20220930", code)
    except :
        error_code_list.append(code)
        continue
    else :
        if df.empty :
            continue
        df['stock_code'] = code
        kospi_stock_trade_volume = pd.concat([kospi_stock_trade_volume, df])
    
kospi_stock_trade_volume.head()
```

```python
kospi_stock_trade_volume.to_csv('./data/kospi_stock_trade_volumn.csv', encoding = 'euc-kr')
```
